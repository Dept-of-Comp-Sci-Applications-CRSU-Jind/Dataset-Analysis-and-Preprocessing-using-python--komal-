Univariate Analysis

Univariate Analysis is a type of data visualization where we visualize only a single variable at a time. Univariate Analysis helps us to analyze the distribution of the variable present in the data so that we can perform further analysis.

```python
fig,ax=plt.subplots ()
plt.figure(figsize =(20,20))
Df ['Infant_deaths'].plot(ax=ax,kind='line', xlabel='Infant_deaths')

ax.set_xlim (0,200)

plt.show()
```

Observation : "infant_deaths" distribution shows the univariat flow of deaths.

```python
fig,ax=plt.subplots ()
plt.figure(figsize =(20,20))
Df ['Life_expectancy'].plot(ax=ax,kind='line', xlabel='Life_expectancy')

ax.set_xlim (1,100)

plt.show()
```

Observation "life_expectancy" : It shows the increasing and decreasing flow of life expectancies.

Principal Component Analysis (PCA) is one of the most commonly used unsupervised machine learning algorithms across a variety of applications: exploratory data analysis, dimensionality reduction, information compression, data de-noising, and plenty more.
PCA can be used to visualize high-dimensional data in two or three dimensions, making it easier to understand and interpret.

PCA

```python
from sklearn import decomposition
```

```python
Pca=decomposition.PCA()
```

```python
Pca.n_components = 0
```

```python
Pca_data=Pca.fit_transform(train_x)
Pca_data
```

```python
Pca.explained_variance_ratio_
```

Feature importance : Feature importance scores are used to determine the relative importance of each feature in a dataset when building a predictive model.
Feature importance is determined by how much each feature contributes to reducing the uncertainty in the target variable. This is typically measured by the amount of reduction in the Gini impurity or entropy that is achieved by splitting on a particular feature.

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit (x,y)
```

```python
importance =model.coef_
importance
```

```python
import matplotlib.pyplot as plt

for i,v in enumerate  (importance ):
  print  ('Features :%0d,Score:%5f' % (i , v))

plt.bar ([x for x in range (len (importance))], height =50)
plt.show ()
```
